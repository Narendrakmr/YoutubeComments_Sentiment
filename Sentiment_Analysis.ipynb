```python
from textblob import TextBlob
```

```python
TextBlob('Say hi to Kong and maverick for me').sentiment.polarity
```

```python
l=[]
for i in df['comment_text']:
    try:
        l.append(TextBlob(i).sentiment.polarity)
    except:
        l.append(0)
    
    
```

```python
len(l)
```

```python
df['polarity']=l
```

```python
positive=df[df['polarity']==1]
```

```python
positive
```

```python
negative=df[df['polarity']== -1]
```

```python
negative
```

```python
negative_comments=' '.join(negative['comment_text'])
```

```python
negative_comments
```

```python
Positive_comments=' '.join(positive['comment_text'])
```

```python
Positive_comments
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
word_cloud=WordCloud(stopwords=set(STOPWORDS)).generate(negative_comments)
plt.figure(figsize=(10,10))
plt.imshow(word_cloud)
plt.axis(False)
```

```python
word_cloud=WordCloud(stopwords=set(STOPWORDS)).generate(Positive_comments)
plt.figure(figsize=(10,10))
plt.imshow(word_cloud)
plt.axis(False)
```

```python
c=df1['comment_text']
```

```python
c
```

```python
l=[]
for i in c:
    for j in i:
        if j in emoji.distinct_emoji_list(i):
            l.append(j)
```

```python
l
```

```python
dict1 = {}
for i in l:
    if i in dict1:
        dict1[i] += 1
    else: dict1[i] = 1
```

```python
obj=list(dict1.keys())
freq=list(dict1.values())
```

```python
import plotly.graph_objs as go
```

```python
from plotly.offline import iplot
```

```python
file=go.Bar(x=obj,y=freq)
```

```python
file1=go.Pie(labels=obj,values=freq)
```

```python
iplot([file1])
iplot([file])
```
